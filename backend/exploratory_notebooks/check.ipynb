```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\AliAsh\Sync\TimeSeriesProject\TimeSearies_toolkit-\backend\data\raw_data\task.csv') 
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.info()
```

```python
df.sort_values(by='timestamp', inplace=True)

# # Plotting the data
# # plt.figure(figsize=(10, 6))
# plt.plot(df['timestamp'], df['value'], color='b')
# plt.xlabel('Timestamp')
# plt.ylabel('Value')
# plt.title('Timestamp vs Value')
# plt.xticks(rotation=45)
# plt.tight_layout()

# # Show the plot
# plt.show()

# Group data by engine_id and plot each group
plt.figure(figsize=(10, 6))
for engine_id, group in df.groupby('engine_id'):
    plt.plot(group['timestamp'], group['value'], label=f'Engine {engine_id}')

plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.title('Timestamp vs Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()

# Show the plot
plt.show()
```
